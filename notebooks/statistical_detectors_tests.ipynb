# Expected flagging rate of non-LLM outlier detectors

This notebook tests the eight individual statistical detectors.  For this test, we disable LLMs and run WP4 with all but one of the detectors detectors enabled.

It is meant as an evaluation notebook only. It does not change the FastAPI service code.

```python
import pandas as pd
import os
for filename in os.listdir('./detector_outputs/'):
    if filename.endswith('.csv'):
        df = pd.read_csv(os.path.join('./detector_outputs/', filename))
        number_of_outliers = df["outlier_detected"].sum()
        print(f"{filename[:-4]}: {number_of_outliers / 109297 * 100:.2f}% of rows are outliers")
```
