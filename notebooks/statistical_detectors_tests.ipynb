# Flagging rate of non-LLM outlier detectors

This notebook tests the eight individual statistical detectors.  For this test first we trained the two trainable detectors on the full `Belge_aus_D` dataset (109297 rows). Then we disable LLMs and run WP4 with all but one of the detectors detectors enabled.  The `./detector_outputs/` folder contains, for each individual detector, exactly those rows which are flagged by that corresponding detector.

It is meant as an evaluation notebook only. It does not change the FastAPI service code.

```python
import pandas as pd
import os
for filename in os.listdir('./detector_outputs/'):
    if filename.endswith('.csv'):
        df = pd.read_csv(os.path.join('./detector_outputs/', filename))
        number_of_outliers = df["outlier_detected"].sum()
        print(f"{filename[:-4]}: {len(df) / 109297 * 100:.2f}% of rows are outliers")
```
