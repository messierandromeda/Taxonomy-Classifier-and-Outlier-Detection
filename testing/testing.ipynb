async def run_version(df, model, prompt, variant, reps):
    frames = []
    for rep in range(reps):
        r = await process_csv(df, model, prompt)
        assert len(r) == len(df), "process_csv changed the row count — can't align by position"
        r['row_id']     = df['row_id'].values
        if TAG: 
            r['difficulty'] = df[TAG].values
        r['variant']    = variant
        r['rep']        = rep
        frames.append(r)
    return pd.concat(frames, ignore_index=True)

versions = [await run_version(df, MODEL, prompt, name, REPS) for name, prompt in VARIANTS.items()]
raw = pd.concat(versions, ignore_index=True)
raw.to_csv(RESULTS, index=False)
